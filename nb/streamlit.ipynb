```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# List of categorical columns to plot
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']

# Number of columns per row
num_cols = 3

# Calculate number of rows needed
num_rows = 3

# Load data
data = pd.read_csv('../dat/feature_eng_data.csv')

# Define a function to create a plot for a given column
def create_plot(col):
    # Group by conversion and the column and calculate count and percentage
    grouped = data.groupby(['treatment_tag', col]).size().reset_index(name='counts')

    # Create stacked bar chart with percentage
    fig, ax = plt.subplots(figsize=(10, 5))
    grouped['percentage'].plot.bar(stacked=False, legend=True, ax=ax)

    # Set title, x-axis label, and y-axis label
    ax.set_title(f'{col} and treatment with percentage')
    ax.set_xlabel(col)
    ax.set_ylabel('Percentage', labelpad=15)

    # Add percentage label on top of each bar
    for p in ax.containers:
        ax.bar_label(p, label_type='edge', labels=[f'{int(round(val))}%' for val in p.datavalues], fontsize=8)

    return fig

# Define the Streamlit app
def app():
    # Set the page title and icon
    st.set_page_config(page_title='Categorical Columns Plot', page_icon=':bar_chart:')

    # Add a title page
    st.title('Categorical Columns Plot')
    st.write('This app plots the distribution of categorical columns by treatment tag.')

    # Add a dropdown to select the column
    selected_col = st.selectbox('Select a column', categorical_cols)

    # Create the plot and show it
    fig = create_plot(selected_col)
    st.pyplot(fig)

if __name__ == '__main__':
    app()
```
